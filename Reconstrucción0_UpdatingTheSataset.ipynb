**DATOS EXTRAS**
1. La funci칩n yf.pdr_override(): te permite proporcionar una funci칩n personalizada que se utilizar치 para descargar datos de Yahoo Finance en lugar de la funci칩n predeterminada proporcionada por pandas_datareader.
```python
# Obtenga datos de Yahoo Finance como dataframe (stickers)
# Se actualiza a datos actuales
yf.pdr_override() # <== Esto es todo lo que se necesita
raw_data  = pdr.get_data_yahoo('^GSPC ^GDAXI ^N225 ^FTSE', start='2018-12-01', end='2023-09-11')
#Visualizar los datos
raw_data
```

**DATOS EXTRAS**
1. pd.concat: Es una funci칩n de la biblioteca pandas en Python que se utiliza para combinar (concatenar) objetos de datos, como DataFrames o Series, a lo largo de un eje espec칤fico. Esta funci칩n es especialmente 칰til cuando trabajas con conjuntos de datos que necesitas unir vertical u horizontalmente para formar un nuevo conjunto de datos m치s grande.

**DATOS EXTRAS**
1. *.copy*: Para copiar una base de datos.

```python
# Crear una copia de seguridad en caso de que eliminemos/alteremos elementos de los datos por error
df_comp = raw_data.copy()
# Verifique las primeras 2 y las 칰ltimas 2 filas
pd.concat([df_comp .head(2), df_comp .tail(2)])
```

```python
#Para ver todos los titulos de las primeras filas
print(df_comp.columns)
```

```python
# Para mantener solo la columna (Adj Close = Precios de cierre)
df_comp = df_comp[['Adj Close']]
df_comp
```

**DATOS EXTRAS**
1. *df_comp = df_comp.asfreq('b')*: se utiliza en pandas en Python para cambiar la frecuencia (frecuencia de muestreo) de un DataFrame de series temporales a d칤as laborables (d칤as h치biles).Cuando trabajas con series temporales en pandas, es com칰n tener datos con diferentes frecuencias temporales, como datos diarios, semanales o mensuales. A veces, es necesario restringir o ajustar la frecuencia de los datos a un calendario espec칤fico. En este caso, 'b' representa "business day frequency", que se refiere a d칤as laborables, excluyendo los fines de semana.

2. *df_comp = df_comp.fillna(method='ffill')*:Se utiliza en pandas en Python para llenar (rellenar) los valores faltantes (NaN o valores nulos) en un DataFrame utilizando un m칠todo espec칤fico, en este caso, el m칠todo de "forward fill" o relleno hacia adelante.

```python
# Configuraci칩n de la frecuencia de los datos.
df_comp=df_comp.asfreq('b')
# Completando los valores faltantes
df_comp=df_comp.fillna(method='ffill')
df_comp
```

**DATOS EXTRAS**
1. *df_comp.rename()*:Se utiliza en pandas en Python para cambiar los nombres de columnas o 칤ndices de un DataFrame. Permite renombrar columnas o 칤ndices existentes seg칰n tus necesidades o preferencias.
```python
# Renombrar  columns to the data set
df_comp= df_comp.rename(columns={"^GDAXI": " dax ", "^GSPC": "spx ", "^N225": "nikkei", "^FTSE": "ftse"})
df_comp
```
