**DATOS EXTRAS**
1. La función yf.pdr_override(): te permite proporcionar una función personalizada que se utilizará para descargar datos de Yahoo Finance en lugar de la función predeterminada proporcionada por pandas_datareader. Esto puede ser útil si deseas ajustar o personalizar la forma en que se recuperan los datos de Yahoo Finance para satisfacer tus necesidades específicas.

```python
# Obtenga datos de Yahoo Finance como dataframe (stickers)
# Se actualiza a datos actuales
yf.pdr_override() # <== Esto es todo lo que se necesita
raw_data  = pdr.get_data_yahoo('^GSPC ^GDAXI ^N225 ^FTSE', start='2018-12-01', end='2023-09-11')
#Visualizar los datos
raw_data
```

**DATOS EXTRAS**
1. pd.concat: Es una función de la biblioteca pandas en Python que se utiliza para combinar (concatenar) objetos de datos, como DataFrames o Series, a lo largo de un eje específico. Esta función es especialmente útil cuando trabajas con conjuntos de datos que necesitas unir vertical u horizontalmente para formar un nuevo conjunto de datos más grande.

```python
# Verifique las primeras 3 y las últimas 3 filas
pd.concat([raw_data .head(3), raw_data .tail(3)])
```

**DATOS EXTRAS**
1. *.copy*: Para copiar una base de datos.

```python
# Crear una copia de seguridad en caso de que eliminemos/alteremos elementos de los datos por error
df_comp = raw_data.copy()
# Verifique las primeras 2 y las últimas 2 filas
pd.concat([df_comp .head(2), df_comp .tail(2)])
```

```python
#Para ver todos los titulos de las primeras filas
print(df_comp.columns)
```

```python
# Para mantener solo la columna (Adj Close = Precios de cierre)
df_comp = df_comp[['Adj Close']]
df_comp
```

**DATOS EXTRAS**
1. *df_comp = df_comp.asfreq('b')*: se utiliza en pandas en Python para cambiar la frecuencia (frecuencia de muestreo) de un DataFrame de series temporales a días laborables (días hábiles).Cuando trabajas con series temporales en pandas, es común tener datos con diferentes frecuencias temporales, como datos diarios, semanales o mensuales. A veces, es necesario restringir o ajustar la frecuencia de los datos a un calendario específico. En este caso, 'b' representa "business day frequency", que se refiere a días laborables, excluyendo los fines de semana.

2. *df_comp = df_comp.fillna(method='ffill')*:Se utiliza en pandas en Python para llenar (rellenar) los valores faltantes (NaN o valores nulos) en un DataFrame utilizando un método específico, en este caso, el método de "forward fill" o relleno hacia adelante.

```python
# Configuración de la frecuencia de los datos.
df_comp=df_comp.asfreq('b')
# Completando los valores faltantes
df_comp=df_comp.fillna(method='ffill')
df_comp
```

**DATOS EXTRAS**
1. *df_comp.rename()*:Se utiliza en pandas en Python para cambiar los nombres de columnas o índices de un DataFrame. Permite renombrar columnas o índices existentes según tus necesidades o preferencias. Puedes utilizarla para hacer que los nombres de las columnas o los índices sean más descriptivos o para cambiarlos por completo.
2. FTSE: London
3. N225: Japon
4. GSPC: EEUU
5. GDAXI: Alemania





```python
# Renombrar  columns to the data set
df_comp= df_comp.rename(columns={"^GDAXI": " dax ", "^GSPC": "spx ", "^N225": "nikkei", "^FTSE": "ftse"})
df_comp
```

```python
#Mostrar los primeros 5 elementos para asegurarse de que los datos se hayan extraído correctamente
print (df_comp.head())
#Asegurarnos de que el último día que incluimos en la serie sea correcto
print (df_comp.tail())
```
