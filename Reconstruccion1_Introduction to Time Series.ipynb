### The QQ Plot

```python
# Importar el módulo scipy.stats para realizar operaciones estadísticas y análisis de datos.
import scipy.stats
# Importar el módulo pylab para crear gráficos y visualizaciones de datos.
import pylab
```

```python
# Crear un QQ Plot (Quantile-Quantile Plot) utilizando scipy.stats.probplot
# para visualizar la distribución de los datos de 'spx ' con respecto a una distribución teórica.
scipy.stats.probplot(df_comp['spx '], plot = pylab)
# Establecer el título del gráfico.
plt.title("QQ Plot", size = 24)
# Mostrar el gráfico.
pylab.show()
```
