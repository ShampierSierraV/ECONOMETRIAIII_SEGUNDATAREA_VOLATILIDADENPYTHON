#**Cargando los datos**

```python
raw_data = yfinance.download(tickers = "^GSPC ^FTSE ^N225 ^GDAXI", start = "1994-01-07", end = "2019-09-01", interval = "1d", group_by = 'ticker', auto_adjust = True, threads = True)
```

El fragmento de código que proporcionaste parece ser parte de la manipulación de un DataFrame llamado df_comp, en el cual se están agregando columnas adicionales que contienen datos de cierre de varios índices bursátiles. Aquí está una descripción de lo que se está haciendo en este código:


*   df_comp['spx'] = df_comp['^GSPC'].Close[:]: Se está creando una nueva columna en el DataFrame df_comp llamada "spx". Los valores en esta columna provienen de la columna "Close" del índice bursátil "^GSPC", lo que significa que está almacenando los precios de cierre del índice S&P 500 en esta columna.
*   df_comp['dax'] = df_comp['^GDAXI'].Close[:]: Se está creando una nueva columna llamada "dax" en el DataFrame df_comp. Al igual que en el primer caso, esta columna almacena los precios de cierre, pero en este caso, son del índice bursátil alemán DAX (^GDAXI).
* df_comp['ftse'] = df_comp['^FTSE'].Close[:]: Se está creando una columna llamada "ftse" en el DataFrame df_comp, que almacena los precios de cierre del índice FTSE 100 (^FTSE) del mercado bursátil del Reino Unido.
* df_comp['nikkei'] = df_comp['^N225'].Close[:]: Finalmente, se crea una columna llamada "nikkei" en el DataFrame df_comp, que almacena los precios de cierre del índice Nikkei 225 (^N225) de Japón.










```python
#Se realiza la copia de seguridad de la base de datos
df_comp = raw_data.copy()
```

```python
df_comp['spx'] = df_comp['^GSPC'].Close[:]
df_comp['dax'] = df_comp['^GDAXI'].Close[:]
df_comp['ftse'] = df_comp['^FTSE'].Close[:]
df_comp['nikkei'] = df_comp['^N225'].Close[:]
```

```python
df_comp = df_comp.iloc[1:]
del df_comp['^N225']
del df_comp['^GSPC']
del df_comp['^GDAXI']
del df_comp['^FTSE']
df_comp=df_comp.asfreq('b')
df_comp=df_comp.fillna(method='ffill')
```

#**Creando devoluciones**
Este fragmento de código está calculando las tasas de rendimiento diarias de los índices bursátiles en el DataFrame df_comp y almacenándolas en nuevas columnas con los nombres "ret_spx," "ret_ftse," "ret_dax," y "ret_nikkei." Aquí se describe lo que se está haciendo en cada línea:

*   df_comp['ret_spx'] = df_comp.spx.pct_change(1).mul(100): Se está calculando la tasa de rendimiento diario del índice S&P 500 (spx) utilizando el método .pct_change(1). Este método calcula el cambio porcentual con respecto al día anterior. Luego, se multiplica por 100 para expresar la tasa de rendimiento en términos porcentuales. El resultado se almacena en la columna "ret_spx."
*   df_comp['ret_ftse'] = df_comp.ftse.pct_change(1).mul(100): De manera similar a la primera línea, se calcula la tasa de rendimiento diario del índice FTSE 100 (ftse) y se almacena en la columna "ret_ftse."
* df_comp['ret_dax'] = df_comp.dax.pct_change(1).mul(100): Aquí se calcula la tasa de rendimiento diario del índice DAX (dax) y se almacena en la columna "ret_dax."
* df_comp['ret_nikkei'] = df_comp.nikkei.pct_change(1).mul(100): De manera análoga a las líneas anteriores, se calcula la tasa de rendimiento diario del índice Nikkei 225 (nikkei) y se almacena en la columna "ret_nikkei."



```python
df_comp['ret_spx'] = df_comp.spx.pct_change(1).mul(100)
df_comp['ret_ftse'] = df_comp.ftse.pct_change(1).mul(100)
df_comp['ret_dax'] = df_comp.dax.pct_change(1).mul(100)
df_comp['ret_nikkei'] = df_comp.nikkei.pct_change(1).mul(100)
```

```python
df_comp['norm_ret_spx'] = df_comp.ret_spx.div(df_comp.ret_spx[1])*100
df_comp['norm_ret_ftse'] = df_comp.ret_ftse.div(df_comp.ret_ftse[1])*100
df_comp['norm_ret_dax'] = df_comp.ret_dax.div(df_comp.ret_dax[1])*100
df_comp['norm_ret_nikkei'] = df_comp.ret_nikkei.div(df_comp.ret_nikkei[1])*100
```

#**ARMAX**

Un modelo ARMAX es una extensión del modelo ARMA (Modelo Autorregresivo de Media Móvil) que incorpora variables exógenas en el modelo. Las variables exógenas son variables independientes que se consideran como factores adicionales que pueden influir en la serie de tiempo que se está modelando. El término "ARMAX" es un acrónimo de "Autorregresivo (AR) de Orden p, Media Móvil (MA) de Orden q, con Variables Exógenas (X)".

```python
model_ret_armax = SARIMAX(df.ret_ftse[1:], exog = df[["ret_spx","ret_dax","ret_nikkei"]][1:], order = (1,0,1))
results_ret_armax = model_ret_armax.fit()

df_pred_armax = results_ret_armax.predict(start = start_date, end = end_date,
                                          exog = df_test[["ret_spx","ret_dax","ret_nikkei"]][start_date:end_date])

df_pred_armax[start_date:end_date].plot(figsize = (20,5), color = "red")
df_test.ret_ftse[start_date:end_date].plot(color = "blue")
plt.title("Predicciones versus reales (devoluciones)", size = 24)
plt.show()
```

#**Pronóstico de la volatilidad**
El pronóstico de la volatilidad se refiere a la estimación o predicción de la futura variabilidad o dispersión de los precios o retornos de un activo financiero, como una acción, un índice bursátil o una moneda. La volatilidad es una medida estadística que cuantifica la magnitud de los cambios en los precios o rendimientos de un activo durante un período de tiempo específico. Es una medida importante en los mercados financieros y es relevante para inversores, operadores y analistas.

La volatilidad puede expresarse en términos absolutos o porcentuales y se utiliza comúnmente para evaluar el riesgo asociado con un activo o una cartera de inversiones. Un activo financiero con una alta volatilidad experimentará fluctuaciones significativas en su precio, lo que lo hace más arriesgado, pero también puede presentar oportunidades de ganancias. Por otro lado, un activo con una baja volatilidad tiende a experimentar movimientos de precios más estables y es visto como menos riesgoso, pero puede ofrecer menos oportunidades de ganancias significativas.

```python
mod_garch = arch_model(df_comp.ret_ftse[1:], vol = "GARCH", p = 1, q = 1, mean = "constant", dist = "Normal")
res_garch = mod_garch.fit(last_obs = start_date, update_freq = 10)
```

```python
pred_garch = res_garch.forecast(horizon = 1, align = 'target')
```

```python
pred_garch.residual_variance[start_date:].plot(figsize = (20,5), color = "red", zorder = 2)
df_test.ret_ftse.abs().plot(color = "blue", zorder = 1)
plt.title("Predicciones de volatilidad", size = 24)
plt.show()
```

#**Modelo de regresión multivariante**
Un modelo de regresión multivariante, también conocido como regresión múltiple, es una técnica estadística utilizada para analizar la relación entre una variable de respuesta (variable dependiente) y múltiples variables predictoras (variables independientes). A diferencia de la regresión simple, que se utiliza para analizar la relación entre una variable de respuesta y una sola variable predictora, la regresión multivariante permite tener en cuenta varias variables predictoras al mismo tiempo.

```python
from statsmodels.tsa.api import VAR
```

```python
df_ret = df[['ret_spx', 'ret_dax', 'ret_ftse', 'ret_nikkei']][1:]
```

```python
model_var_ret = VAR(df_ret)
model_var_ret.select_order(20)
results_var_ret = model_var_ret.fit(ic = 'aic')
```

```python
results_var_ret.summary()
```

```python
lag_order_ret = results_var_ret.k_ar
var_pred_ret = results_var_ret.forecast(df_ret.values[-lag_order_ret:], len(df_test[start_date:end_date]))

df_ret_pred = pd.DataFrame(data = var_pred_ret, index = df_test[start_date:end_date].index,
                                columns = df_test[start_date:end_date].columns[4:8])

df_ret_pred.ret_nikkei[start_date:end_date].plot(figsize = (20,5), color = "red")

df_test.ret_nikkei[start_date:end_date].plot(color = "blue")
plt.title("Real vs Predicción", size = 24)
plt.show()
```
